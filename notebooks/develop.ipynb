## Tournament

```python
tour = Tournament(
    players=[Always0(Game()), Always3(Game()), UniformRandom(Game()), Focal5(Game()), TitForTat(Game()), CastigadorInfernal(Game())],
    n_rounds=100,
    error=0.05,
    repetitions=2,
)
```

## Evolution

```python
evo = Evolution(
    players=[Always0(Game()), Always3(Game()), UniformRandom(Game()), Focal5(Game()), TitForTat(Game()), CastigadorInfernal(Game())],
    n_rounds=100,
    error=0.05,
    generations=2,
)
```

```python
df_evo = evo.play_trace()
```
